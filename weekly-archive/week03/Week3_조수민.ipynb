# Week 1: 인과추론 입문

## 개요
- **연관 ≠ 인과**  
  단순히 같이 움직이는 변수들이 있다고 해서 원인과 결과 관계는 아님.  
- **인과추론의 핵심 질문**  
  “만약 개입을 했더라면(What if), 결과가 어떻게 달라졌을까?”  
- **근본문제(Fundamental Problem of Causal Inference)**  
  한 개인에 대해 ‘처치 시 결과’와 ‘비처치 시 결과’를 동시에 관측할 수 없다.  
- 따라서 설계, 가정, 추론 전략이 필요하다.  

---

## 핵심 개념
- **처치(Treatment, T)**: 우리가 개입하려는 변수 (예: 광고 이메일 발송 여부)  
- **결과(Outcome, Y)**: 관심 있는 성과 (예: 구매 여부)  
- **잠재결과(Potential Outcomes)**: 처치 받았을 때 결과, 혹은 처치 받지 않았을 때 결과  
- **인과효과(Causal Effect)**: 처치 받았을 때 결과 - 처치 받지 않았을 때 결과  

---

## 근본문제
- 실제로는 각 개인에 대해 하나의 결과만 관측됨.  
- 다른 하나는 반사실(counterfactual).  
- 인과추론은 이 반사실을 추정하기 위한 틀과 방법론을 제공한다.  

---

## 식별(Identification)과 가정
- **교환가능성(Exchangeability)**: 처치 여부가 결과와 독립이라는 가정  
  - 무작위 배정 실험에서, 처치가 들어가기 전 까지 두 집단은 거의 동질한 집단이어야 함  
- **SUTVA** (Stable Unit Treatment Value Assumption)  
  - 간섭 없음 (한 사람의 처치가 다른 사람 결과에 영향 X)  
  - 일관성 (처치 정의가 모호하지 않음)  
    - SUTVA가 깨지는 경우 예시 (신약 효과성 실험 시)
      - 예: 원래 실험 프로토콜은 약을 하루에 한 알 먹는것이지만, 실험군 참가자 A가 약 먹는 것을 깜빡하여 이틀에 한 알씩 먹음
      - 이 경우 A는 다른 실험군들과 동등한 처치를 받은 것이 아니므로, 분석에서 제외하여야 함
      

---

## Selection Bias (선택 편향)
- 처치 집단과 통제 집단이 본질적으로 다른 경우 발생  
- 예: 젊은 사람만 신약을 복용하고, 노인들은 복용하지 않은 경우  
- 대조군이 반사실을 제대로 대변하지 못해 인과효과 추정이 왜곡됨

# Week 2: 무작위 실험과 기초 통계

## 개요
- 무작위 배정(Randomization)은 교란을 제거하고 인과효과 식별을 가능케 하는 가장 좋은 방법
- A/B 테스트를 통해 RCT를 실무에 적용 가능
---

## 왜 무작위화인가?
- 교환가능성(혹은 독립성)을 가장 확실하게 보장.  
- 실험군과 대조군이 사전에 동질하다는 가정을 확보하면, 처치 이후의 단순 집단 평균 비교가 곧 인과효과 추정이 됨.  

---

## 가설 검정을 위한한 기본 통계
- **추정량의 분산(Variance of Estimator)**: 표본 평균의 분산은 모집단 분산/표본 크기. n이 커질수록 분산이 줄고 추정이 안정 -> 대수의 법칙, 중심극한정리. n수는 대체로 클수록 좋다
- **표준오차(SE)**: 표본 추정치의 불확실성, 분산의 제곱근.
- **신뢰구간(CI)**: 추정치 ± (임계값 × SE)로 불확실성을 수치 범위로 표현.
- **가설검정 & p값**: 효과가 없다는 가정 하에서 관측값이 얼마나 극단적인가.
- **검정력(Power) & 표본 크기**: 실제 효과를 잡아낼 수 있는 능력과 설계 조건.
- **효과크기(Effect Size)**: 통계적 유의성과 별개로 실제 차이가 얼마나 큰지 설명해주는 지표.
 

# 질문 | 혹은 아이디어 낙서장

Non-inferiority Test (비열등성 검정) 관련

## Non-inferiority Test란?
- 비열등성 검정은 기존 처치(대조군) 대비 새로운 처치(실험군)가 **유의미한 수준으로 차이**나지 않는다(열등하지 않다)는 것을 증명하기 위한 통계적 방법.
- 주로 **의학/임상시험**에서 신약이나 새로운 치료법을 기존 표준 치료와 비교할 때 사용된다.

---

## 검정 구조
- **귀무가설(H0)**: 새로운 처치는 기존 대비 Δ 이상 열등하다.  
- **대립가설(H1)**: 새로운 처치는 기존 대비 Δ 이상 열등하지 않다.  
- Δ = 비열등성 한계(non-inferiority margin) 는 주로 경험적, 이론적으로 정의됨 (예: Δ = 10%p면, 효과 차이가 10% 정도 차이나는거는 용인)  

---
